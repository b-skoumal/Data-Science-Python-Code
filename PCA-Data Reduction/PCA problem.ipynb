# Import libraries

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import scale
from sklearn.decomposition import PCA
```

# Load data

```python
data = pd.read_csv('b5.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
#convert the data to a numpy array
X = data.values
```

```python
X
```

## Scale the values

```python
#scale the values so that all the variables are in the same range 
#(and so that one variable doesn't overwhelm the others)
X = scale(X)
```

```python
X
```

# Conduct PCA
## PCA with 50 components

```python
#Do the actual PCA analysis, with 50 components, since we have 50 variables. 
#This shows how much of the total variance each one of these components explains
#Starts with 1 and goes to 40
#The first accounts for 16.1 percent of the variance, the second accounts for 9.2 percent, etc.
pca = PCA()
pca.fit(X)
print(pca.explained_variance_ratio_)
```

## PCA with 5 components

```python
#since there are supposedly 5 "personality factors" the data should reflect a 5-component solution
#so we will do PCA for 5 components
pca5 = PCA(n_components=5)
pca5.fit(X)
print(pca5.explained_variance_ratio_)
```

## Contributions of variables to components

```python
#so for each of the 50 variables, we see how much each variable contributes to each of the components, 
print(pca5.components_)
```
